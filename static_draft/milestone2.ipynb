```python
# AI 
# I could not figure out how to find mark_() encoding for lollipop and I asked chatGPT ang gave it my progress.
year_columns = [col for col in sector.columns if col.startswith('Y_')]

# Melt the dataframe to get years in a single column
melted_data = pd.melt(sector, id_vars=['Country_code_A3', 'Substance'], 
                      value_vars=year_columns, 
                      var_name='Year', value_name='Emissions')

# Convert Year to numeric, removing 'Y_' prefix
melted_data['Year'] = pd.to_numeric(melted_data['Year'].str.replace('Y_', ''))

# Calculate total emissions for each year
yearly_totals = melted_data.groupby('Year')['Emissions'].sum().reset_index()

# Create the lollipop chart
base = alt.Chart(yearly_totals).encode(
    x=alt.X('Year:O', axis=alt.Axis(title='Year', labelAngle=45)),
    y=alt.Y('Emissions:Q', axis =alt.Axis( title = 'Emissions (Kilotons)'))
).properties(
    title = "Total GHG Emissions in the World (1990 - 2018)"
)

lines = base.mark_rule().encode(
    y='Emissions:Q'
)

points = base.mark_circle(size=100).encode(
    y='Emissions:Q',
    color=alt.value('steelblue')
)


# Combine all layers
final_chart = (lines + points).configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_title(
    fontSize=16
)

# Display the chart
final_chart
```

## This chart will show that the countries' increase in emissions have been really large over the last ~10 years. I don't want this chart to be misleading, so it might be a good idea to sort this by population as well or maybe use absolute changes as they are more important in climate change discussions.

```python
mean_emissions = sector.groupby(['Name', 'dev_country'])[['Y_2010', 'Y_2018']].mean().reset_index()
mean_emissions.head()

mean_emissions = mean_emissions[mean_emissions['dev_country'] != 0]
mean_emissions['dev_country'] = mean_emissions['dev_country'].replace('D', 'Developing')
mean_emissions['dev_country'] = mean_emissions['dev_country'].replace('I', 'Industrialized')

mean_emissions['percentage_change'] = ((mean_emissions['Y_2018'] - mean_emissions['Y_2010']) / mean_emissions['Y_2010']) * 100
mean_emissions.replace([float('inf'), -float('inf')], pd.NA, inplace=True)

mean_emissions['percentage_change'] = pd.to_numeric(mean_emissions['percentage_change'], errors='coerce')

mean_emissions.dropna(subset=['percentage_change'], inplace=True)

top_10_countries = mean_emissions.nlargest(10, 'percentage_change').sort_values(by='percentage_change', ascending=False)

alt.Chart(top_10_countries).mark_bar().encode(
    x=alt.X('percentage_change:Q', title='Percentage Change in Emissions (2010 to 2018)'),
    y=alt.Y('Name:N', title='Country'),
    color='dev_country:N'
).properties(
    title='Top 10 Countries by Percentage Change in Emissions (2010-2018)',
    width=600,
    height=300
)
```
